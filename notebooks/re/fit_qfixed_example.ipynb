A simple example of how to fit numeric resource esitimation with a polynomial.

```python
from qmath.utils.re_utils import re_numeric_fixed_point
from qmath.func.common import AbsInPlace
import numpy as np

small_to_zero = lambda a: np.array([0 if abs(x)<1e-9 else x for x in a])

op = AbsInPlace()
n_range = range(5,30)
re = [re_numeric_fixed_point(op, {"n": n, "radix":n//2}, n_inputs=1) for n in n_range]
for metric in ["gidney_lelbows", "gidney_relbows", "toffs","qubit_highwater", "active_volume"]:
    print(metric, small_to_zero(np.polyfit(n_range, [x[metric] for x in re], deg=3)))
```
