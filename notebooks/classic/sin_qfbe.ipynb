```python
# Evaluating cos(pi*x), x in [0,1). 
# Reference: https://arxiv.org/pdf/2001.00807

import numpy as np

# Computes cos(pi * x * 2^-r).
def cos_qfbe(x, r):
    assert 0 <= x < 2**r
    v = [(x>>i)%2 for i in range(r)]
    a=1
    for i in range(0, r):
        v_prev = 0 if i==0 else v[i-1]
        s = 1 if (v[i]==v_prev) else -1
        a = ((1+s*a)/2)**0.5    
    return a if v[r-1]==0 else -a

r=20
for x in range(2**r):
    val1 = cos_qfbe(x,r)
    val2 = np.cos(np.pi * x * 2**(-r))
    assert np.isclose(val1, val2)
print("OK")
```
