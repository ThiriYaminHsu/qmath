Bug repro: multiplying QFixed by classical real number doesn't work. 

```python
from psiqworkbench import QPU, QFixed
import psiqworkbench.qubricks as qbk

qpu = QPU(filters=[">>64bit>>", ">>bit-sim>>"])
qpu.reset(500)
qx = QFixed(40,name="qx", radix=20, qpu=qpu)
qy = QFixed(40,name="qy", radix=20, qpu=qpu)
qz = QFixed(40,name="qz", radix=20, qpu=qpu)

# QFixed by QFixed fixed works.
qx.write(1.25)
qy.write(2.75)
qz.write(0)
qbk.GidneyMultiplyAdd().compute(qz, qx, qy)
print("OK:", qz.read())

# QFixed by classical integer works.
qx.write(1.25)
qz.write(0)
qbk.GidneyMultiplyAdd().compute(qz, qx, 2)
print("OK:", qz.read())
```

```python
# QFixed by classical real number - doesn't work.
qx.write(1.25)
qz.write(0)
qbk.GidneyMultiplyAdd().compute(qz, qx, 2.75)
print("OK:", qz.read())
```
